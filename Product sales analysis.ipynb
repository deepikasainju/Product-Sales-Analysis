```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv('Order_details(masked).csv')
dataset
```

```python
df = dataset.copy()
```

```python
df.isnull().sum()
```

```python
df.info()
```

# Converting transaction date to datetime and hours

```python
df['Time'] = pd.to_datetime(df['Transaction Date'])
df['Hour'] = (df['Time']).dt.hour
df
```

# finding n busiest hours where n= 30

```python
# n = 30 in this case, can be modified 
# as per need to see top 'n' busiest hours 
timemost1 = df['Hour'].value_counts().index.tolist()[:30]  
timemost2 = df['Hour'].value_counts().values.tolist()[:30]
```

timemost2 = number of times the hours occur

```python
timemost = np.column_stack((timemost1,timemost2)) 
print(" Hour Of Day" + "\t" + "Cumulative Number of Purchases \n") 
print('\n'.join('\t\t'.join(map(str, row)) for row in timemost)) 
```

```python
tmost = df['Hour'].value_counts() 
timemost1 = [] 
  
for i in range(0,23): 
    timemost1.append(i) 
      
timemost2 = tmost.sort_index() 
timemost2.tolist() 
timemost2 = pd.DataFrame(timemost2) 
```

```python
timemost2
```

```python
plt.figure(figsize=(20, 10)) 
plt.title('Sales Happening Per Hour (Spread Throughout The Week)') 
plt.ylabel("Number Of Purchases Made") 
plt.xlabel("Hour") 
plt.plot(timemost1, timemost2) 
plt.grid() 
plt.show() 
```

# Observation

The most sales is occuring in hour 12 and 23

```python
import seaborn as sns
```

```python
df['Product'] = df['Product'].str.split(',')
df_exploded = df.explode('Product')
```

# The explode function in pandas is used to transform each element of a list-like column (e.g., a column containing lists, tuples, or arrays) into a separate row, while replicating the other columns in the DataFrame.

```python
product_counts = df_exploded['Product'].value_counts().reset_index()
product_counts.columns = ['Product', 'Count']
product_counts
```

```python
plt.figure(figsize=(20, 6))
sns.barplot(data=product_counts, x='Product', y='Count', palette='viridis')
plt.xlabel('Product')
plt.ylabel('Number of People')
plt.title('Number of People Selecting Each Product')
plt.show()
```

# Observation
Product_75 is the most popular product
